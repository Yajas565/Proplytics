## Multivariate analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\capstone_project_dscience\real_estate_project\dataset\cleaned_gurgaon_v2.csv").drop_duplicates()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

**property_type  vs  price**

```python
sns.barplot(x=df['property_type'], y=df['price'], estimator='median')
```

```python
sns.boxplot(x=df['property_type'], y=df['price'])
```

**property_type  vs  area**

```python
sns.barplot(x=df['property_type'], y=df['area'], estimator='median')
```

```python
sns.boxplot(x=df['property_type'], y=df['area'])
```

```python
# removing the outlier
df = df[df['area'] <= 400000]
```

```python
sns.boxplot(x=df['property_type'], y=df['area'])
```

**property_type  vs  price_per_sqft**

```python
sns.barplot(x=df['property_type'], y=df['price_per_sqft'], estimator='median')
```

```python
sns.boxplot(x=df['property_type'], y=df['price_per_sqft'])
```

**property_type  vs  bedRoom**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pd.crosstab(df['property_type'], df['bedRoom']), annot=True, fmt='d')
```

```python
#outliers in bedRoom column
df[df['bedRoom'] >= 10]
```

**property_type  vs  floorNum**

```python
(plt.figure(figsize=(20, 6)))
sns.heatmap(pd.crosstab(df['property_type'], df['floorNum']), annot=True, fmt='d')
```

```python
sns.boxplot(x=df['property_type'], y=df['floorNum'])
```
