## 1.3 cleaning and feature extraction
        property_name column
            1) i want to extract the sector name from the property_name column


```python
df['property_name'].value_counts()
```

     here i extracted the sector detail from the property_name column

```python
sectors = df['property_name'].str.split('in').str.get(1).str.split('Gurgaon').str.get(0).str.strip()
```

```python
df.insert(1, 'sector', sectors)
```

```python
df.head()
```

```python
df['sector'] = df['sector'].str.lower()
```

```python
df['sector'].value_counts()
```

***society column** <br>
        here in society column we can see that the some digits as well as characters are also attached with the names

```python
df['society'].value_counts()
```

```python
df['society'] = df['society'].str.lower()
```

    now lets remove these digits and characters from the names

```python
df['society'] = df['society'].str.replace(r'[\d\★]', '', regex=True)
```

```python
df['society'].value_counts()
```

**price**<br>

```python
df['price'].value_counts()
```

```python
df['price'] = df['price'].str.lower()
```

    the price column is string dtype here. so i converted every value(except price on value and NAN) to crore and make them float type

```python
def price_to_number(price):
    if pd.isna(price):
        return np.nan
    if 'lac' in price:
        return float(price.replace('lac','').strip())/100
    elif 'crore' in price:
        return float(price.replace('crore','').strip())
    else: return price
    

df['price'] = df['price'].apply(price_to_number)
```

```python
df.head()
```

    the rate column name should be price per sq ft instead of rate.
        1) so we will first make remvoe the sq/ft from the values and then make them numerical dtype
        2) also change the name of the column 'rate' -> 'price_per_sqft'


```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.split(' ').str.get(1).str.replace(',','').astype(float)
```

```python
df.head()
```

```python
df['area'] = round((df['price']/df['price_per_sqft'])*10000000,2).astype(float)
```

**bedroom**<br>
    converting bedroom to numerical dtype by removing the Bedrooms string from the string

```python
df['bedRoom'].value_counts()
```

    here i removed the null values where the price column is null(i have to remove these rows because area values for these prices are also NULL, so i was not able to predict the price values)

```python
df = df[~df['price'].isnull()]
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).str.strip().astype(float)
```

**bathroom**<br>
    removing the bathroom keyword from the string

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).str.strip().astype(float)
```

**balcony**<br>
    we can't convert it into a numerical value because there is a value '3+ Balconies'

```python
df['balcony'].value_counts()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.strip()
```

    i am not able to get any useful information from the **address* column so i am going to remove it, but if later on if it is required then i will take the action    

```python
df.drop(columns=['address'], inplace=True)
```
